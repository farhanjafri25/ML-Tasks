**Data Visualization**

```
plt.figure(figsize= (30,30))
df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per sub-category')
plt.show()
```

```
plt.figure(figsize=(12,10))
df['Sub-Category'].value_counts().plot.pie(autopct= '%1.1f%%')
```

```
sns.scatterplot('Sales','Profit',data = df)
```

```
sns.pairplot(df)
```

```
df.plot(kind = 'scatter', x = 'Profit', y = 'Sales', figsize=(14,8), color= 'darkblue')
sns.regplot('Sales', 'Profit', data= df, color= 'red')
plt.show()
```

**From the above Scatter and Regplot plot:**

Profit is not solely depending upon Sales. There are many other factors acting uopn it

```
 df.plot(kind='scatter', x='Quantity', y='Profit', figsize=(12,6), color= 'orange')
 plt.show()
```

**From the above scatter plot:**

Profit is also not depending based upon Quantity. There are many other factors actin upon it.

```
plt.figure(figsize=[12,8])
ax = sns.barplot(x='Region', y= 'Sales',hue='Segment', data= df, palette= 'Blues')
```

```
pd.crosstab(df['Quantity'], df['Category'], df['Profit'], aggfunc='sum').plot(kind='bar',stacked=True)
```
